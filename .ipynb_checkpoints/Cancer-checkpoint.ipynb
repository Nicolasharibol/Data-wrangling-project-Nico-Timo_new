```python
# Locate the table with population counts over the years
table2 = soup2.find('table', {'class': 'table table-striped table-bordered table-hover table-condensed table-list'})
#table2 (to check if it works)
```

```python
# Extract population data from the table
population_data_chile = []
for row in table2.find('tbody').find_all('tr'):
    cells = row.find_all('td')
    if len(cells) > 1:  # Ensure it's a valid row with data
        year = cells[0].text.strip()  # Year column
        population = cells[1].text.strip()  # Population column
        population_data_chile.append((year, population))
# Print the extracted data
for year, population in population_data_chile: #### created population data chile
    print(f"Year: {year}, Population: {population}")
```

```python
# Locate the table with population counts over the years
table3 = soup3.find('table', {'class': 'table table-striped table-bordered table-hover table-condensed table-list'})
#table3 (to check if it works)
```

```python
# Extract population data from the table
population_data_germany = []
for row in table3.find('tbody').find_all('tr'):
    cells = row.find_all('td')
    if len(cells) > 1:  # Ensure it's a valid row with data
        year = cells[0].text.strip()  # Year column
        population = cells[1].text.strip()  # Population column
        population_data_germany.append((year, population))
# Print the extracted data
for year, population in population_data_germany:# created population data germany 
    print(f"Year: {year}, Population: {population}")
```

```python
# Create graphs to illustrate the answers to the hypotheses.
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# make a graph illustrating the total cancer incidence over time for Germany

# Set the aesthetic style of the plots
sns.set_theme(style="whitegrid")

# Create the line plot
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='total_cancer_incidence', data=df_germany_cancer_population_total_5_short, marker='o', linewidth=5)

# Set the range of x-axis and y-axis
plt.xlim(1989, 2016)  # x-axis range: 1989 to 2016
plt.ylim(0.2, 0.3)     # y-axis range: 0.2 to 0.3

# Adjust the font size of the x-axis and y-axis tick labels
plt.xticks(fontsize=16)  # Set font size for x-axis values
plt.yticks(fontsize=12)  # Set font size for y-axis values

# Add labels and a title
plt.title("Development of Cancer Incidences in Germany over Time", fontsize=20)
plt.xlabel("Time [Year]", fontsize=16)
plt.ylabel("Cancer Incidences in total Population [%]", fontsize=16)

# Safe the plot (needs to be done first, before showing it!)
plt.savefig("cancer_trend_germany_plot.png", dpi=300)

# Show the plot
plt.show()
```

```python
# make a graph illustrating the breast cancer incidence over time for Germany

# Set Seaborn style
sns.set_theme(style="whitegrid")

# Create a figure and an axis
fig, ax1 = plt.subplots(figsize=(10, 6))

# Create the histogram for cancer cases
bar_width = 3.5  # The width of the bars; lower values result in larger gaps
ax1.hist(df_germany_population_cancer_breast['Year'], bins=len(df_germany_population_cancer_breast['Year']), weights=df_germany_population_cancer_breast['Breast cancer '], color='darkblue', alpha=0.7, edgecolor='black', width=bar_width)

# Set labels for the first y-axis
ax1.set_ylabel("Number of Breast Cancer Cases", fontsize=18, color='darkblue')
ax1.set_xlabel("Time [Year]", fontsize=18)
ax1.tick_params(axis='y', labelcolor='darkblue', labelsize=14) # Set font size for y-axis values

# Adjust the font size of the x-axis labels
plt.xticks(fontsize=16)  # Set font size for x-axis values

# Set the range of x-axis and y-axis
plt.xlim(1989, 2016)  # x-axis range: 1989 to 2016
plt.ylim(0, 40000)     # y-axis range: 0 to 40000

# Create a second y-axis for the line plot
ax2 = ax1.twinx()
sns.lineplot(x='Year', y='Cancer_Percentage', data=df_germany_population_cancer_breast, ax=ax2, marker='o', color='red', linewidth=5)

# Set labels for the second y-axis
ax2.set_ylabel("Cancer Incidences in total Population [%]", fontsize=18, color='red')
ax2.tick_params(axis='y', labelcolor='red', labelsize=14) # Set font size for y-axis values

# Set the range of y-axis
plt.ylim(0.02, 0.03)     # y-axis range: 0.02 to 0.03

# Add a title
plt.title("Development of Breast Cancer Incidences in Germany over Time", fontsize=18)

# Safe the plot (needs to be done first, before showing it!)
plt.savefig("breast_cancer_trend_germany_plot.png", dpi=300)

# Adjust layout for better appearance
plt.tight_layout()

# Show the plot
plt.show()
```
