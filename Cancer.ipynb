```python
# Locate the table with population counts over the years
table2 = soup2.find('table', {'class': 'table table-striped table-bordered table-hover table-condensed table-list'})
#table2 (to check if it works)
```

```python
# Extract population data from the table
population_data_chile = []
for row in table2.find('tbody').find_all('tr'):
    cells = row.find_all('td')
    if len(cells) > 1:  # Ensure it's a valid row with data
        year = cells[0].text.strip()  # Year column
        population = cells[1].text.strip()  # Population column
        population_data_chile.append((year, population))
# Print the extracted data
for year, population in population_data_chile: #### created population data chile
    print(f"Year: {year}, Population: {population}")
```

```python
# Extract population data from the table
population_data_germany = []
for row in table3.find('tbody').find_all('tr'):
    cells = row.find_all('td')
    if len(cells) > 1:  # Ensure it's a valid row with data
        year = cells[0].text.strip()  # Year column
        population = cells[1].text.strip()  # Population column
        population_data_germany.append((year, population))
# Print the extracted data
for year, population in population_data_germany:# created population data germany 
    print(f"Year: {year}, Population: {population}")
```
